```python
import pandas as pd
```

```python
# Read in file
deaths_data = pd.read_csv("../Resources/causes_of_deaths.csv")
deaths_data.head()
```

```python
# Columns selected
death_columns_selected = deaths_data[['Unnamed: 0', 'Abortion', 'Embolism', 'Haemorrhage', 
                                      'Hypertension', 'Sepsis', 'Other direct causes', 'Indirect causes']].copy()
death_columns_selected.head()
```

```python
# Remove nan row (index 0)
death_selected_rows = death_columns_selected.loc[1. :].copy()
death_selected_rows
```

```python
# Remaned columns
death_selected_rows.columns = map(str.lower, death_selected_rows.columns)

death_renamed = death_selected_rows.rename(columns={'unnamed: 0': 'region', 
                                                     'other direct causes': 'other_direct_causes',
                                                    'indirect causes': 'indirect_causes'}).copy()

death_renamed.head()
```

```python
death_renamed.dtypes
```

```python
death_dtype_modified = death_renamed.astype({'abortion': 'int64', 'embolism': 'int64',
                                            'haemorrhage': 'int64', 'hypertension': 'int64',
                                            'sepsis': 'int64', 'other_direct_causes': 'int64',
                                            'indirect_causes': 'int64'}).copy()
```

```python
death_renamed.columns
```

```python
# Reset index
death_reset_index = death_renamed.reset_index().copy()
death_reset_index = death_reset_index.rename(columns={'index': 'id'})
death_reset_index.head()
```

```python
# csv filepath
death_completed_csv = '../output_file/causes_of_deaths_final.csv'

death_reset_index.to_csv(death_completed_csv, index=False)
```

### Load to Database

```python
from sqlalchemy import create_engine
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy import Column, Integer, String, Float
from sqlalchemy.orm import Session

from config import username, password, host, database
```

```python
# Sets an object to utilize the default declarative base in SQL Alchemy
Base = declarative_base()
```

```python
# Create table schema
class Causes(Base):
    __tablename__ = 'causes_of_deaths'
    __table_args__ = {'extend_existing': True}
    id = Column(Integer, primary_key=True)
    region = Column(String(255))
    abortion = Column(Integer)
    embolism = Column(Integer)
    haemorrhage = Column(Integer)
    hypertension = Column(Integer)
    sepsis = Column(Integer)
    other_direct_causes = Column(Integer)
    indirect_causes = Column(Integer)
```

```python
# Create Engine
connect_string = f'postgresql://{username}:{password}@{host}/{database}'
engine = create_engine(connect_string)
```

```python
# Specify table
causes_table = [Base.metadata.tables['causes_of_deaths']]

# Create (if not already in existence) the tables associated with our classes.
Base.metadata.create_all(engine, tables=causes_table)
```

```python
# Store data to SQL database
death_reset_index.to_sql(name='causes_of_deaths', con=engine, if_exists='append', index=False)
```

```python
# Check of the table exists in database
causes_sql = pd.read_sql_query('select * from causes_of_deaths', con=engine)
causes_sql.head()
```
